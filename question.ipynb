### 1. What percentage of passengers survived?

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('titanic_fake.csv')

# Show first 5 rows
df.head()

# Calculate survival percentage
survival_rate = df['Survived'].mean() * 100
print(f"Survival Rate: {survival_rate:.2f}%")

```

### 2. Survival Rate by Gender

```python
# Group by Sex and calculate mean survival
gender_survival = df.groupby('Sex')['Survived'].mean() * 100
print("Survival Rate by Gender:\n", gender_survival)
```

### 3. Average age of survivors and non survivors

```python
# Group by Survived and calculate average age
avg_age = df.groupby('Survived')['Age'].mean()
print("Average Age:\n", avg_age)
```

### 4. Plot number of passengers by class

```python
import matplotlib.pyplot as plt

# Count by Pclass
df['Pclass'].value_counts().sort_index().plot(kind='bar')
plt.title('Passenger Count by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

### 5. Visualize age distribution

```python
import seaborn as sns

# Histogram of age
sns.histplot(df['Age'].dropna(), kde=True, bins=30, color='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```
